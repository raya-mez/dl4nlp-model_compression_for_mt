```python
import json
import pickle
import pandas as pd
from matplotlib import pyplot as plt
```

```python
language_pairs = ['fr_FR', 'de_DE', 'it_IT', 'hi_IN', 'sw_KE', 'th_TH']
```

```python
LANGUAGE_BY_CODE = {
    "ar_EG": "Arabic","ar_SA": "Arabic","bg_BG": "Bulgarian","bn_BD": "Bengali","bn_IN": "Bengali",
    "ca_ES": "Catalan","cs_CZ": "Czech","da_DK": "Danish","de_DE": "German","el_GR": "Greek",
    "es_MX": "Spanish","et_EE": "Estonian","fa_IR": "Farsi","fi_FI": "Finnish","fil_PH": "Filipino",
    "fr_CA": "French","fr_FR": "French","gu_IN": "Gujarati","he_IL": "Hebrew","hi_IN": "Hindi",
    "hr_HR": "Croatian","hu_HU": "Hungarian","id_ID": "Indonesian","is_IS": "Icelandic","it_IT": "Italian",
    "ja_JP": "Japanese","kn_IN": "Kannada","ko_KR": "Korean","lt_LT": "Lithuanian","lv_LV": "Latvian",
    "ml_IN": "Malayalam","mr_IN": "Marathi","nl_NL": "Dutch","no_NO": "Norwegian","pa_IN": "Punjabi",
    "pl_PL": "Polish","pt_BR": "Portuguese","pt_PT": "Portuguese","ro_RO": "Romanian","ru_RU": "Russia",
    "sk_SK": "Slovak","sl_SI": "Slovenian","sr_RS": "Serbian","sv_SE": "Swedish","sw_KE": "Swahili",
    "sw_TZ": "Swahili","ta_IN": "Tamil","te_IN": "Telugu","th_TH": "Thai","tr_TR": "Turkish",
    "uk_UA": "Ukrainian","ur_PK": "Urdu","vi_VN": "Vietnamese","zh_CN": "Mandarin","zh_TW": "Mandarin",
    "zu_ZA": "Zulu"
}
```

```python
pseudo_result_folder_base_path = '../results/pseudo_quant/'
```

```python
result_dict = {}
for k in range(1, 17):
    result_path = pseudo_result_folder_base_path + f'{k:02d}_bit_model/'
    result_dict[f'{k:02d}'] = {}
    for lang_pair in lang_pairs:
        with open(os.path.join(result_path, f'scores_{lang_pair}.json'), 'r') as file:
            lang_result = json.load(file)
        result_dict[f'{k:02d}'][LANGUAGE_BY_CODE.get(lang_pair)] = {}
        result_dict[f'{k:02d}'][LANGUAGE_BY_CODE.get(lang_pair)]['COMET'] = lang_result['COMET']
        result_dict[f'{k:02d}'][LANGUAGE_BY_CODE.get(lang_pair)]['CHRF++'] = lang_result['CHRF++']
        result_dict[f'{k:02d}'][LANGUAGE_BY_CODE.get(lang_pair)]['BLEU'] = lang_result['BLEU']
```

```python
result_dict.keys()
```

```python
french_commet_scores = []
french_chrf_scores = []
french_bleu_scores = []
german_commet_scores = []
german_chrf_scores = []
german_bleu_scores = []
italian_commet_scores = []
italian_chrf_scores = []
italian_bleu_scores = []
hindi_commet_scores = []
hindi_chrf_scores = []
hindi_bleu_scores = []
swahili_commet_scores = []
swahili_chrf_scores = []
swahili_bleu_scores = []
thai_commet_scores = []
thai_chrf_scores = []
thai_bleu_scores = []

for key in result_dict.keys():
    french_commet_scores.append(result_dict[key]['French']['COMET'])
    french_chrf_scores.append(result_dict[key]['French']['CHRF++'])
    french_bleu_scores.append(result_dict[key]['French']['BLEU'])

    german_commet_scores.append(result_dict[key]['German']['COMET'])
    german_chrf_scores.append(result_dict[key]['German']['CHRF++'])
    german_bleu_scores.append(result_dict[key]['German']['BLEU'])

    italian_commet_scores.append(result_dict[key]['Italian']['COMET'])
    italian_chrf_scores.append(result_dict[key]['Italian']['CHRF++'])
    italian_bleu_scores.append(result_dict[key]['Italian']['BLEU'])

    hindi_commet_scores.append(result_dict[key]['Hindi']['COMET'])
    hindi_chrf_scores.append(result_dict[key]['Hindi']['CHRF++'])
    hindi_bleu_scores.append(result_dict[key]['Hindi']['BLEU'])

    swahili_commet_scores.append(result_dict[key]['Swahili']['COMET'])
    swahili_chrf_scores.append(result_dict[key]['Swahili']['CHRF++'])
    swahili_bleu_scores.append(result_dict[key]['Swahili']['BLEU'])

    thai_commet_scores.append(result_dict[key]['Thai']['COMET'])
    thai_chrf_scores.append(result_dict[key]['Thai']['CHRF++'])
    thai_bleu_scores.append(result_dict[key]['Thai']['BLEU'])
```
